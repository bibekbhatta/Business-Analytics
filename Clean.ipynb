### defining a function to do repetitive task

```python
def Normalize(score):
    if score > 100:
        return score/10
    elif score <= 100:
        return score
```

```python
df['Chemistry'] = df['Chemistry'].apply(Normalize)
```

```python
df['Science'] = (df['Physics']+df['Chemistry']+df['Biology'])/3
```

```python
from sklearn.decomposition import PCA # to reduce data using Principal Component Analysis
```

```python
pca = PCA(n_components=1) # we only plan to use the first principal component out of the four variables
```

```python
pca.fit(df_telecom[['Broadband', 'Telephone', 'Mobile', 'FiveG']])
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_ # shows the proportion of variances captured by the principal component/s
```

```python
x_pca = pca.transform(df_telecom[['Broadband', 'Telephone', 'Mobile', 'FiveG']])
```

```python
x_pca
```

```python
df_telecom["TelePCA1"] = pd.DataFrame(x_pca) # adding x_pca to the existing dataframe df_telecom
```

```python
##to keep or retain only required columns as follows:
df_telecom2 = df_telecom[["Name", "TelePCA1"]]
```
