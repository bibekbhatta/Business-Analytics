## Consolidate /merge the two files 

```python
# to merge the two dfs based on a given column
df_combined = pd.merge(df1, df2, on='Name', how='inner', validate="one_to_one")
```

```python
df = df_combined.copy() # making a separate copy of the combined file
```

```python
df.isnull().sum()
```

```python
##dropping rows having any NaN (null) value
# df = df.dropna()
```

```python
df['Physics'].fillna(value = df['Physics'].mean(), inplace = True)
df['Biology'].fillna(value = df['Biology'].mean(), inplace = True)
```

### defining a function to do repetitive task

```python
def Normalize(score):
    if score > 100:
        return score/10
    elif score <= 100:
        return score
```

```python
df['Chemistry'] = df['Chemistry'].apply(Normalize)
```

```python
df['Science'] = (df['Physics']+df['Chemistry']+df['Biology'])/3
```

### defining another function...

```python
def Dummy_stay(note):
    if note == "Home":
        return 1
    elif note == "Hostel":
        return 0
```

```python
df["Home"] = df["Stays"].apply(Dummy_stay)
```
