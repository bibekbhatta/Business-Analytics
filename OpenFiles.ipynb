# To access multiple text files

```python
import glob
import os
```

```python
path = './DataFile'
```

```python
file_list = glob.glob(path+"/*.txt")
```

```python
file_list
```

```python
corpus = []
```

```python
for file_path in file_list:
    with open(file_path, encoding="utf8") as f_input:
        corpus.append(f_input.read())

```

```python
corpus
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(file_list, columns=['Filenames'])
```

```python
df
```

```python
df["Content"] = pd.DataFrame(corpus)
```

```python
df.head(3)
```

```python
#save file as csv
df.to_csv('Text_combined.csv', index=False)
```

# Access one pdf file

```python
path = './DataFile'
```

```python
## we are using PyPDF2 package in this case; many other alternatives are available; e.g tika
! pip install PyPDF2 #if you are using Anaconda, install it differently without using pip
```

```python
import PyPDF2 
# creating a pdf file object 
pdfFileObj = open(path+"/Microsoft 2021 Annual Report.pdf", "rb") 
```

```python
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
```

```python
# creating a page object 
pageObj = pdfReader.getPage(0) 
print(pageObj.extractText())     
# extracting text from the first (0) page 
```

```python
## access data from all pages 
# printing number of pages in pdf file 
print(pdfReader.numPages) 
all_page = pdfReader.numPages  
#all_page
```

```python
for pages in range(all_page):
    #print(pages)
    pageObj = pdfReader.getPage(pages) 
    #print(pageObj.extractText()) # uncomment if all pages need to be printed below  
```

```python
# closing the pdf file object 
pdfFileObj.close() 
```
