# Access one pdf file

```python
path = '/home/jovyan/DataFile' #path to the file
```

```python
## we are using PyPDF2 package in this case; many other alternatives are available; e.g tika
## More recent version of PyPDF2 is available now; if we install the newest version, the code will have to change slightly
! pip3 install PyPDF2==2.11.1 #if you are using Anaconda, install it differently without using pip. 
```

```python
import PyPDF2 
# creating a pdf file object 
pdfFileObj = open(path+"/Microsoft 2021 Annual Report.pdf", "rb") 
```

```python
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
```

```python
# creating a page object 
pageObj = pdfReader.getPage(0) 
print(pageObj.extractText())     
# extracting text from the first (0) page 
```

```python
## access data from all pages 
# printing number of pages in pdf file 
print(pdfReader.numPages) 
all_page = pdfReader.numPages  
#all_page
```

```python
for pages in range(all_page):
    #print(pages)
    pageObj = pdfReader.getPage(pages) 
    #print(pageObj.extractText()) # uncomment if all pages need to be printed below  
```

```python
# closing the pdf file object 
pdfFileObj.close() 
```

# Accessing web page

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://books.toscrape.com/catalogue/category/books/travel_2/index.html" 
```

```python
response = requests.get(url) # getting the information from webpage
```

```python
soup = BeautifulSoup(response.content, "html.parser") # parsing the extracted text to get required information
```

```python
titles = soup.find_all('h3') # find title
```

```python
prices = soup.find_all('p', class_='price_color') #find price
```

```python
stocks = soup.find_all('p', class_='instock availability') #find availability
```
