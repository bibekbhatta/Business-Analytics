## Linear Regression using statsmodels

```python
!pip install statsmodels # CAREFUL!!! when installingn in local machine, use Anaconda to install in the right environment 
```

```python
import statsmodels.api as sm
```

```python
#define Y variable
y = df['sales']
```

```python
#define explanatory variables
x = df[['TV', 'radio', 'newspaper']]
```

```python
#add constant to predictor variables
x = sm.add_constant(x) # in many packages (e.g Stata, this would be done automatically)
```

```python
#fit linear regression model
model = sm.OLS(y, x).fit()
```

```python
#view model summary
print(model.summary())
```

## Classification

```python
df = pd.read_csv(path+"/Kickstarter.csv",) # CAREFUL! this replaces the existing df if you have one
```

```python
df.head(3) # just to observe the data
```

```python
#define Y variable
y = df['Success']
```

```python
#define explanatory variables
x = df[['word_count', 'Pictures', 'Video', 'backers_count','Lngoal']]
```

```python
!pip install scikit-learn # installing sklearn # CAREFUL!!! when installingn in local machine, use Anaconda to install in the right environment 
```

```python
from sklearn.model_selection import train_test_split # import this as sklearn and NOT scikit-learn
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25) # separating the data for training and testing purpose on 75:25 basis
```

```python
from sklearn.preprocessing import StandardScaler # for the purpose of scaling if needed
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(X_train) # scaling the training set
xtest = sc_x.transform(X_test) # scaling the testing set
```

```python
xtrain[0:2, :] # checking if the variables have been scaled; first two rows and all columns
```

```python
from sklearn.linear_model import LogisticRegression # import from within sklearn
```

```python
classifier = LogisticRegression() # various optionsn available 
```

```python
classifier.fit(X_train, y_train) # fitting the model
```

```python
y_pred = classifier.predict(X_test) # predicting the outcome on test database based on the generated model
```

```python
from sklearn.metrics import confusion_matrix
Con_Mat = confusion_matrix(y_test, y_pred) # to check accuracy based on confusion matrix
```

```python
print ("Confusion Matrix: \n", Con_Mat) # print confusion matrix generated above
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```
